```python
import math
import numpy as np
```

```python
#This is our one-dimentional function
def f(x):
    return 0.5*(6*x-2)**2*math.sin(12*x-4)-10*(x-0.5)-5
```

# 

Base on the above algorithm which is from handout number 3, we can find the upper value of x, so in the while loop below we continue the loop until f(k)<f(k-1) but if f(k-1)<f(k-2) and f(k-1)<f(k) the loop is break and x upper is fined.  

```python
xl = 0.1
delta = 0.02
i = 1
x0 = xl + delta
bound = [xl,x0]

while f(bound[-1])<f(bound[-2]):
    
    x = xl + delta*(1.618)**i
    bound.append(x)
    i = i + 1

    if f(bound[-2])<f(bound[-1]):
        xu = bound[-1]
        xl = bound[-3]
        
        
print(i) 
print(bound)   
```
